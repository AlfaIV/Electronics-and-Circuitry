***Решение операторным методом***

Преобразуем исходную схему к виду, удобному для анализа операторным методом. Результат преобразований представлен на рисунке ниже:



Преобразуем входное напряжение в виде функции:
$U_1(t) = 100[\eta(t) - \eta(t-t_и)]$

Применим преобразование Лапласса:
$U_1(p) = 100(\frac{1}{p} - \frac{e^{-pt_и}}{p}) = 100\frac{1 - e^{-pt_и}}{p}$

Теперь для преобразованной схемы, запишем:

$
\begin{equation*}
 \begin{cases}
   I_1 = I_2 + I_3\\
   U_1 = I_1R_2 + I_2R_3 = I_1R_2 + I_3(\frac{1}{pC} + R_4) - \frac{U_{C}(0)}{p}\\
   U_2 = I_2R_3 = I_3(R_4 + \frac{1}{pC}) - \frac{U_c(0)}{p}
 \end{cases}
\end{equation*}
$

По формулам разброса получим:

$
\begin{equation*}
 \begin{cases}
   I_2 = I_1\frac{R_4 + \frac{1}{pC}}{R3 + R4 + \frac{1}{pC}}\\
   I_3 = I_1\frac{R_3}{R3 + R4 + \frac{1}{pC}}
 \end{cases}
\end{equation*}
$

Учтем то, что общий импеданс цепи равен:
$
Z = R_2 + \frac{R_3 (\frac{1}{pC} + R_4)}{R_3 + \frac{1}{pC} + R_4} = \frac{(R_3 + \frac{1}{pC} + R_4)R_2 + R_3 (\frac{1}{pC} + R_4)}{R_3 + \frac{1}{pC} + R_4} = \frac{pC(R_2R_3 + R_2R_4 + R_3R_4) + R_2 + R_3 }{pC(R_3 + R_4) + 1}
$

Тогда:
$
I_1 = \frac{U_1}{Z}
$

Проведем ряд последовательных преобразований:

$
 U_2 = I_2R_3 = I_1\frac{R_4 + \frac{1}{pC}}{R3 + R4 + \frac{1}{pC}}R_3 = \frac{U_1}{Z}\frac{R_4 + \frac{1}{pC}}{R3 + R4 + \frac{1}{pC}}R_3 = \frac{U_1}{ \frac{pC(R_2R_3 + R_2R_4 + R_3R_4) + R_2 + R_3 }{pC(R_3 + R_4) + 1}}\frac{R_4 + \frac{1}{pC}}{R3 + R4 + \frac{1}{pC}}R_3 =\frac{pCR_4 + 1}{pC(R_2R_3 + R_2R_4 + R_3R_4) + R_2 + R_3 } R_3U_1
$

После обратного преобразования Лапласса получим следущую функцию

```python
from sympy.integrals import laplace_transform, inverse_laplace_transform
U = U_1*(sym.Heaviside(t) - sym.Heaviside(t-t_i))
_U_ = laplace_transform(U,t,p)[0]
#U_2 = U_1/(R_2 + (R_4+1/(p*C))/(R_3 + R_4 + 1/(p*C))*(1/(p*C)+R_4))*R_3/(R_3 + R_4 + 1/(p*C))
#U_2 = U_1*((p*C*R_4*R_3 + R_3)/(p*C*(R_2*R_3 + R_2*R_4 + R_3*R_4)) + R_2 + R_3)
U_2 = (p*C*R_4 + 1)/(p*C*(R_2*R_3 + R_2*R_4 + R_3*R_4) + R_2 + R_3 )*R_3*_U_
U_2 = inverse_laplace_transform(U_2,p,t)
#U_2
from IPython.display import display, Math, Latex
display(Math(r'U_2 = '),U_2)
```

Теперь посторим график этой функции, чтобы сравнить с полученной выше зависимостью.
